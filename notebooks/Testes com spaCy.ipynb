```python
import spacy
from spacy import displacy
from IPython.core.display import HTML
from spacy.util import minibatch, compounding
from spacy.gold import GoldParse
from pathlib import Path
from spacy.scorer import Scorer

import random

import plac
nlp = spacy.load("pt_core_news_sm")
```

```python
doc = nlp("O Rodrigo mora na Rua Adelino 123 de Barueri São Paulo. Nós vamos ganhar essa porra. Estamos tentando identificar a Bacia do Rio Atibaia. Ou até mesmo a Área Indígena Rio Negro Ocaia.")
def print_ents(doc):
  for ent in doc.ents:
      print(ent.text, ' - ', ent.start, ent.end, ent.label_)
print_ents(doc)
```

```python
def plot_dependencies(doc):
  html = displacy.render(doc, style="dep", options={'distance': 100, 'compact': True})
  display(HTML(html))
```

```python
doc = nlp("O Rodrigo mora na Rua Adelino 123 de Barueri São Paulo. Nós vamos ganhar essa porra. Estamos tentando identificar a Bacia do Rio Atibaia. Ou até mesmo a Área Indígena Rio Negro Ocaia.")
plot_dependencies(doc)
```
